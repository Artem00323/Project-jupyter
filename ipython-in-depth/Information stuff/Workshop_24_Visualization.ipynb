```python
data.head()
```

```python
data.Fare.value_counts()
```

```python
data.Fare.apply(lambda x: x // 5 * 5  if x  < 100 else 100)
```

```python
data['fare'] = data.Fare.apply(lambda x: x // 5 * 5  if x  < 100 else 100)
```

```python
ex = data[['Survived','SibSp','fare']].groupby(['SibSp','fare'],as_index=False).sum()
ex.sort_values(by='Survived',ascending=False).head()
```

```python
pt = pd.pivot_table(ex,values='Survived', index=['SibSp'],columns=['fare']).fillna(0)
sns.clustermap(pt)
```
